```python
####################

import os
import base64
import sys

from google.colab import output


ROOT = "/content"
rep_download_name = "fast_dataset_maker"
encoded_data = b'aHR0cHM6Ly9naXRodWIuY29tL1dTSDAzMi9zZC13ZWJ1aS1mYXN0LWRhdGFzZXQtbWFrZXIuZ2l0'
rep_url = base64.b64decode(encoded_data).decode('utf-8')
rep_dir = os.path.join(ROOT, rep_download_name)
extensions_dir_name = "extensions"
extensions_dir = os.path.join(rep_dir, extensions_dir_name)
requirments_versions_txt_path = os.path.join(rep_dir, "requirements_versions.txt")
requirments_txt_path = os.path.join(rep_dir, "requirements.txt")
deduplicate_cluster_dir = os.path.join(
    extensions_dir,
    "image_deduplicate_cluster_webui"
)


########## 上面的常量声明，请从第一个代码块复制过来，以确保重启后重新初始化 ##########

#@title ## 🛠️ 1.2 Set config | 配置环境


#@markdown **只要重启过内核就要再次运行此代码块**\
#@markdown **只要重启过内核就要再次运行此代码块**\
#@markdown **只要重启过内核就要再次运行此代码块**


#@markdown ---

#@markdown 强制重写config
force_rewrite = True #@param {type:"boolean"}

#@markdown **1.2.1 infinite_image_browsing设置**

#@markdown `infinite_image_browsing` 中 `文生图`,`图生图` 对应的路径
outdir_txt2img_samples = "/content/images" #@param {type:"string"}
outdir_img2img_samples = "/content/images" #@param {type:"string"}

#@markdown **1.2.2 dataset_tag_editor设置**

#@markdown `dataset_tag_editor` 中显示的图片分辨，低分辨率可以加快显示，网络较差时请选择`128 / 64 `
max_resolution = 192 #@param {type:"number"}


####################
import json
import logging
import textwrap
from typing import List

os.chdir(ROOT)


# 自检
extensions_name_list = [
    "dataset_tag_editor_standalone",
    "Gelbooru_API_Downloader",
    "image_deduplicate_cluster_webui",
    "sd_webui_infinite_image_browsing",
]

def check_extensions_dir(extensions_name_list: List[str], extensions_dir: str):
    if not extensions_dir:
        return

    exist_extensions_folder_name_list = os.listdir(extensions_dir)
    not_exist_extensions_name_list = [
        extension
        for extension in extensions_name_list
        if extension not in exist_extensions_folder_name_list
    ]
    if not_exist_extensions_name_list:
        logging.warning(
            textwrap.dedent(f"""\
                以下扩展不存在于{extensions_dir}\n{not_exist_extensions_name_list}
                请检查 1.1 中是否正常Clone， 或者提issue: {rep_url}
            """)
        )
check_extensions_dir(extensions_name_list, extensions_dir)


# 扩展rep路径
if rep_dir not in sys.path:
    sys.path = [rep_dir] + sys.path


# 设置 tag_editor 配置
def set_tag_editor_config(force_rewrite: bool):
    tag_editor_config_dict =dict(
        use_temp_files = True,
        temp_directory = "",
        cleanup_tmpdir = False,
        image_columns = 6,
        max_resolution = max_resolution,
        filename_word_regex = "",
        tagger_use_spaces = True,
        interrogator_use_cpu = False,
        interrogator_keep_in_memory = False,
        interrogator_max_length = 30,
        interrogator_model_dir = "",
    )
    tag_editor_config_path = os.path.join(
        extensions_dir,
        "dataset_tag_editor_standalone",
        "settings.json",

    )
    # 用户可能会在WebUI里自行设置，如果没开强制重写，这里不要重置了
    if not os.path.exists(tag_editor_config_path) or force_rewrite:
        with open(tag_editor_config_path, "w") as f:
            json.dump(tag_editor_config_dict, f)
set_tag_editor_config(force_rewrite)


# 设置 image_browsing_config
sd_webui_config_path = os.path.join(
    extensions_dir,
    "sd_webui_infinite_image_browsing",
    "config.json",
)
def set_image_browsing_config():
    sd_webui_config_dict = dict(
        outdir_txt2img_samples = outdir_txt2img_samples,
        outdir_img2img_samples = outdir_img2img_samples,
        # outdir_extras_samples = "",
    )
    # 用户可能会在WebUI里自行设置，如果没开强制重写，这里不要重置了
    if not os.path.exists(sd_webui_config_path) or force_rewrite:
        with open(sd_webui_config_path, "w") as f:
            json.dump(sd_webui_config_dict, f)
set_image_browsing_config()


# 导入 downloader
downloader_dir = os.path.join(
    extensions_dir,
    "Gelbooru_API_Downloader",
)
sys.path = [downloader_dir] + sys.path
```

```python
#@title ### ⏬ Download images | 下载图片

#@markdown 你可以在这找到tags规则 | some useful info for tags

#@markdown - [tags](https://gelbooru.com/index.php?page=wiki&s=&s=view&id=25921)

#@markdown - [cheatsheet](https://gelbooru.com/index.php?page=wiki&s=&s=view&id=26263)

#@markdown ---

#@markdown 符合gelbooru搜索规则的tags
tags = "1girl" #@param {type:"string"}

#@markdown 需要下载的 `图片数`, `下载路径`, `下载单位`
max_images_number = 100 #@param {type:"integer"}
download_dir = "/content/images" #@param {type:"string"}
unit = 100 #@param {type:"slider", min:1, max:100, step:1}

#@markdown 与图片一起被下载的txt中tag规则：`逗号分割`，`移除下划线`，`对括号进行转义`
add_comma = True #@param {type:"boolean"}
remove_underscore = True #@param {type:"boolean"}
use_escape = True #@param {type:"boolean"}

#@markdown 对下载的图片进行完整性检查，以避免出现截断图片

#@markdown | -1 | 0 | 1 | 2 |
#@markdown |------|------|------|------|
#@markdown | 不检查 | 检查但只输出信息 | 检查并修复 | 检查并删除 |


check_images_mode = -1 #@param {type:"slider", min:-1, max:2, step:1}


from download_images_coroutine import Scrape_images

check_images_mode = None if check_images_mode == -1 else 1

Scrape_images_kwargs = dict(
    tags = tags,
    max_images_number = max_images_number,
    max_workers = 10,
    download_dir = download_dir,
    unit = unit,
    add_comma = add_comma,
    remove_underscore = remove_underscore,
    use_escape = use_escape,
    check_images_mode = check_images_mode,
    )
await Scrape_images(**Scrape_images_kwargs)
```
